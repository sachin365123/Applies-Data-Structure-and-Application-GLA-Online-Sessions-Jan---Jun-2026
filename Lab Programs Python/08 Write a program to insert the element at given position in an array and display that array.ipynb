## Write a program to insert the element at given position in an array and display that array. 

```python
# Python program to insert an element at a given position in an array

# Taking input from user
arr = []

n = int(input("Enter number of elements in array: "))
```

```python
print("Enter array elements:")
for i in range(n):
    element = int(input())
    arr.append(element)
```

```python
# Element to insert
new_element = int(input("Enter element to insert: "))
```

```python
# Position where element should be inserted
position = int(input("Enter position to insert element (0-based index): "))
```

```python
# Inserting element
arr.insert(position, new_element)
```

```python
# Display updated array
print("Array after insertion:")
print(arr)
```
