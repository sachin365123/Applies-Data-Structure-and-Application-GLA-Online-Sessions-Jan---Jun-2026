## 09 Write a program to delete the element from given position in an array and display that array. 

```python
# Python program to delete an element from a given position in an array
# without using pop()

arr = []

n = int(input("Enter number of elements in array: "))

print("Enter array elements:")
for i in range(n):
    element = int(input())
    arr.append(element)
```

```python
# Position to delete
position = int(input("Enter position to delete element (0-based index): "))
```

```python
# Check valid position
if position >= 0 and position < n:

    # Shift elements to left
    for i in range(position, n - 1):
        arr[i] = arr[i + 1]

    # Remove last duplicate element
    del arr[n - 1]

    print("Array after deletion:")
    print(arr)

else:
    print("Invalid Position")
```
